```python
def get_data(fname, prefix):
    personas = []
    s = len(prefix)
    cur = []
    with open(fname,'r') as f:
        for line in f:
            # print(line)
            line = ' '.join(line.strip().split()[1:]) # remove line number
            if line[:s] == prefix:
                cur.append(line[s:].strip())
            else:
                if len(cur)>0:
                    personas.append(cur)
                    cur = []
    return personas
```
